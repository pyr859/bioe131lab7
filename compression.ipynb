```python
# import packages
import numpy as np
from Bio import Entrez
from Bio import SeqIO
Entrez.email = 'y.peng@berkeley.edu'
```

```python
# generate six random binary files that contain 50% to 100% of zeros; write to files
prob = 1
for i in range(6):
    data = np.random.choice([0, 1], size=8*1024*1024*100, replace=True, p=[prob, 1-prob])
    data = np.packbits(data)
    open("zeros_" + str(prob) + "p", "wb").write(data)
    prob -= 0.1

print(prob)
```

```python
# generate a random nucleotide sequence and a random amino acid sequence; write to files
nt = np.random.choice(["A", "T", "C", "G"], size=100000000, replace=True)
aa = np.random.choice(['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y'],
                     size=100000000, replace=True)
open("nt_seq.fa", "w").write("".join(nt))
open("aa_seq.fa", "w").write("".join(aa))
```

## Q: What is the difference between bzip2 and pbzip2? Do you expect one to be faster and why?
### A: pbzip2 is a parallel implementation of the bzip2 block-sorting file compressor that uses multiple cores in a processor. I expect pbzip2 to be faster.

## Q: How does the level of compression change as the percentage of zeros increases? Why does this happen?
### A: The level of compression increases as the percentage of zeros inceases. bzip2 and pbzip2 use Borrows-Wheeler transform to increase compression efficiency for the run-length encoding. Run-length encoding works by representing a series of zeros with two values: a zero followed by an integer that denotes the length of zeros.

```python
# Perform an Entrez search and fetch to acquire the FASTA sequences of 
# the envelope glycoprotein gp120 of HIV virus.
seqs = []
search = Entrez.esearch(db = 'nuccore',
                        term = 'envelope glycoprotein AND "Human immunodeficiency virus"[porgn:__txid12721] AND 2000:5000[sequence length]',
                        sort = 'relevance',
                        idtype = 'acc',
                        retstart = 0,
                        retmax = 10)
for i in Entrez.read(search)['IdList']:
    gene_fetch = Entrez.efetch(db = 'nuccore', id = i, rettype = 'fasta', retmode = 'text')
    record = SeqIO.read(gene_fetch, 'fasta')
    seqs.append(record)

with open("gp120.fasta", "w") as output_handle:
    SeqIO.write(seqs, output_handle, "fasta")
    
print(seqs)
```
